```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data Frame
df = pd.read_csv("student_result.csv")
print(df.head())
```

```python
df.describe()                      # used to generate descriptive statistics of a DataFrame
```

```python
df.info()                         # Index Range, Column Names, Non-null Counts, Data Types, Memory Usage
```

```python
# Count the no. of NULL values:
df.isnull().sum()
```

# Drop unnamed column:

```python
df = df.drop("Unnamed: 0", axis = 1)
```

```python
print(df.head())
```

# Change weekly study hours column:

```python
# df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5 - 10", "5-10")
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct", "5 - 10")
df.head()
```

# Gender Distribution:

```python
# Count the Values:
plt.figure(figsize = (5, 5))                               # {width, height}
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
# From the above chart we have analyzed that
# The Number of Females in the Data is more than the Number of Males
```

# Impact of Parent Education on Different Score:

```python
# Group By:
gb = df.groupby("ParentEduc").agg({ "MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean' })
print(gb)
```

```python
sns.heatmap(gb)
```

```python
# To show the values in the heat map:
sns.heatmap(gb, annot = True)
plt.show()
```

```python
# To Decrease heatmap size in Width and Height:
plt.figure(figsize = (4,4))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parents Eduction and Student's Score")
plt.show()
```

```python
# From the above chart we have concluded that the education of the parents have a good impact on their Scores
```

# Group by Parent Marital Status:

```python
gb1 = df.groupby("ParentMaritalStatus").agg({ "MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean' })
print(gb1)
```

```python
# Seaborn Chart:
plt.figure(figsize = (4,4))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that there is no/negLigibLe impact on the
#student 's score due to their parent's meital status
```

# Ethinic Gorup Distributions:

```python
print(df["EthnicGroup"].unique())
```

```python
# Percentage Distribution => PIE Chart:
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["group A", "group B", "group C", "group D", "group E"]
myList = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]

print("My List: ", myList)
plt.pie(myList, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
